```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

### Load Data

```python
data = pd.read_csv("txt_files/tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

### Lowercase

```python
data['lowercase_review'] = data['Review'].str.lower()
```

```python
data.head()
```

### Stop word removal

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove("not")
```

```python
data['review_no_stopwords'] = data['lowercase_review'].apply(lambda x: ' '.join(word for word in x.split() if word not in (en_stopwords)))
```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

### Punctuation

We want to remove punctation but we notice we have a few reviews with the symbol * instead of the word star. We want to keep this as it might add meaning to the review. We can do this using what we learned about regular expressions.

```python
data['review_no_stopword_no_punct'] = data.apply(lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']), axis=1)
```

```python
data.head()
```

```python
data['review_no_stopword_no_punct'] = data.apply(lambda x: re.sub(r"([^\w\s])", "", x['review_no_stopword_no_punct']), axis=1)
```

```python
data.head()
```

### Tokenizing

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopword_no_punct']), axis=1)
```

```python
data['tokenized'][0]
```

### Stemming

```python
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

### Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens]) 
```

```python
data['lemmatized'][0]
```

```python
data.head()
```

### N-grams

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
```

```python
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```

```python
ngrams_4 = (pd.Series(nltk.ngrams(tokens_clean, 4)).value_counts())
print(ngrams_4)
```
